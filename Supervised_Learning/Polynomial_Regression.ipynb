```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
train_data = pd.read_csv('poly_data.csv')
train_data
```

```python
X = train_data['Var_X'];
print(X)
print(X.values) # values gives numpy array
```

```python
X = X.values.reshape(-1, 1)
print(X)
print(X.shape)
```

```python
y = train_data['Var_Y'].values;
y
```

```python
poly_feat = PolynomialFeatures(degree = 4)
X_poly = poly_feat.fit_transform(X)
X_poly
# so for degree 4, for every x it returns x^0,x^1,x^2,x^3,x^4
```

```python
# Now we fit above x values for every y
poly_model = LinearRegression(fit_intercept = False).fit(X_poly,y)
poly_model
```

```python
prediction=poly_model.predict(np.array([[1,-1.01925,1.03887056,-1.05886882,1.07925205]]))
prediction
```

```python
# Give 100 x co-ordinates spacing from X.min() to X.max()
X_plot=np.linspace(X.min(),X.max(),100).reshape(-1,1)
X_plot
```

```python
X_plot_poly=poly_feat.fit_transform(X_plot)
X_plot_poly 
# observer that first value, 1, 3, 9, 27, 81
# so for degree 4 - x^4, x^3, x^2, x^1
```

```python
# plot the results
import matplotlib.pyplot as plt

# first plot with X, y 
plt.plot(X,y,"bo")

# Give 100 x co-ordinates spacing from -3 to +3 
X_plot=np.linspace(X.min(),X.max(),100).reshape(-1,1)
# Applying poly feat on above x plots
X_plot_poly=poly_feat.fit_transform(X_plot)

# using above X_plot and prediction of X_plot_poly plot
plt.plot(X_plot,poly_model.predict(X_plot_poly),'-b')
plt.show()
```
