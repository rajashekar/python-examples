```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
```

```python
train_data = pd.read_csv('regularization_data.csv')
train_data.values
```

```python
X = train_data.values[:,:-1]
X
```

```python
y = train_data.values[:,-1]
y
```

```python
lasso_reg = Lasso()
lasso_reg.fit(X,y);
```

```python
reg_coef = lasso_reg.coef_
reg_coef
```

In above we see that, first and last co-efficents was 0, so first and last features are removed <br>
Now we scale feature and check how it behaves <br>

```python
scaler = StandardScaler()
scaler
```

```python
X_Scaled = scaler.fit_transform(X)
X_Scaled
```

```python
lasso_reg.fit(X_Scaled,y);
reg_coef = lasso_reg.coef_
reg_coef
```
